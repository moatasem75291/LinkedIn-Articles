```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
# generate some Linear-looking data to test the equation
np.random.seed(42)
x = 2 * np.random.rand(100, 1)
y = 4 + 3 * x + np.random.rand(100, 1)
```

```python
plt.scatter(x,y)
plt.xlabel("X1");plt.ylabel("Y");
```

```python
# Compute theta using normal equation => (X^t X)^-1 (X^t y) 
x_b = np.c_[np.ones_like(x), x]
theta_best = np.linalg.inv(x_b.T.dot(x_b)).dot(x_b.T.dot(y))
```

```python
theta_best
```

```python
# Make prediction using theta hat
X_new = np.array([
    [0], [2]
])
X_new_b = np.c_[np.ones_like(X_new), X_new]
y_predict = X_new_b.dot(theta_best)
```

```python
y_predict
```

```python
# plot this model predictions
# plt.figure(figsize=(5,3))
plt.plot(X_new, y_predict, "r-")
plt.plot(x, y, "b.")
plt.axis([0, 2, 0, 15])
plt.show()
```

```python
# Performing Linear Regression using Sklearn.
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(x, y)
```
